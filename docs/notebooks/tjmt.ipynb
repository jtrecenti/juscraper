# TJMT - Tribunal de Justica do Estado de Mato Grosso

Scraper for the jurisprudence search of TJMT.

| Feature | Status |
|---------|--------|
| cjsg (Acordao) | Supported |
| cjsg (Decisao Monocratica) | Supported |
| cpopg | Not implemented |
| cposg | Not implemented |

## Basic search

```python
import juscraper as jus

tjmt = jus.scraper('tjmt')

dados_cjsg = tjmt.cjsg('dano moral', paginas=1)
print(dados_cjsg.shape)
dados_cjsg.head(3)
```

## Available columns

```python
dados_cjsg.columns.tolist()
```

## Preview ementa

```python
print(dados_cjsg['ementa'].iloc[0][:300])
```

## Using filters

You can filter by date range, judge (relator), court chamber (orgao_julgador), and case type (tipo_processo).

```python
dados_filtrados = tjmt.cjsg(
    'direito',
    paginas=1,
    data_julgamento_inicio='2025-01-01',
    data_julgamento_fim='2025-06-30',
    tipo_processo='Criminal',
)
print(dados_filtrados.shape)
dados_filtrados.head(3)
```

## Querying different types

Besides `Acordao` (default), you can also query `DecisaoMonocratica`.

```python
dados_decisao = tjmt.cjsg('direito', paginas=1, tipo_consulta='DecisaoMonocratica')
print(dados_decisao.shape)
dados_decisao.head(3)
```

## Download and parse separately

You can also download the raw data first and parse it later.

```python
brutos = tjmt.cjsg_download('direito', paginas=1)
print(f"Pages downloaded: {len(brutos)}")
print(f"Keys in response: {list(brutos[0].keys())}")

dados = tjmt.cjsg_parse(brutos)
import pandas as pd
df = pd.DataFrame(dados)
print(f"\nParsed shape: {df.shape}")
df.head(3)
```

# TJGO — Court of Justice of Goiás

Scraper for the TJGO jurisprudence search hosted on Projudi
(`ConsultaJurisprudencia`). The page renders a Cloudflare Turnstile
widget but the backend does not validate the token, so the scraper
works with pure HTTP requests.

```python
import juscraper as jus

tjgo = jus.scraper('tjgo')
```

## Basic search

Download two pages of results so we can verify pagination works.

```python
df = tjgo.cjsg('dano moral', paginas=range(1, 3))
print(df.shape)
df.head(3)
```

## Available columns

```python
df.columns.tolist()
```

## Text preview

```python
print(df['texto'].iloc[0][:300])
```

## Filtering by instance and area

The search form exposes dropdowns that we can map directly:

* ``id_instancia``: ``0`` all / ``1`` 1st degree / ``2`` recursal / ``3`` tribunal
* ``id_area``: ``0`` all / ``1`` civil / ``2`` criminal

```python
df_civel = tjgo.cjsg('dano moral', paginas=1, id_instancia=3, id_area=1)
df_civel[['processo', 'serventia', 'tipo_ato']].head()
```

## Download and parse separately

`cjsg_download` returns the raw HTML responses, which is useful when
you want to keep the original payloads.

```python
raw = tjgo.cjsg_download('dano moral', paginas=1)
print(type(raw), len(raw), 'chars in page 1:', len(raw[0]))
df_raw = tjgo.cjsg_parse(raw)
df_raw.head(3)
```

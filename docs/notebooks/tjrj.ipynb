# TJRJ — Court of Justice of Rio de Janeiro

Scraper for the TJRJ jurisprudence search (`ConsultarJurisprudencia.aspx`).
The site displays a reCAPTCHA widget but the backend does not validate it,
so the scraper works with pure HTTP requests.

```python
import juscraper as jus

tjrj = jus.scraper('tjrj')
```

## Basic search

Download two pages of results so we can verify pagination works.

```python
df = tjrj.cjsg('dano moral', ano_inicio=2024, ano_fim=2024, paginas=range(1, 3))
print(df.shape)
df.head(3)
```

## Download and parse separately

`cjsg_download` returns the raw JSON payloads from the server — useful when
you want to keep the original data.

```python
raw = tjrj.cjsg_download('dano moral', ano_inicio=2024, ano_fim=2024, paginas=1)
print(type(raw), len(raw))
print('Total de documentos na base:', raw[0]['TotalDocs'])
df_raw = tjrj.cjsg_parse(raw)
df_raw.head(3)
```

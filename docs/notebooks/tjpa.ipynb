# TJPA - Tribunal de Justica do Estado do Para

Scraper for the jurisprudence search of the Court of Justice of the State of Para (TJPA).

| Feature | Available |
|---------|-----------|
| cjsg    | Yes       |
| cpopg   | No        |
| cposg   | No        |

## Basic search

```python
import juscraper as jus

# Create the TJPA scraper
tjpa = jus.scraper('tjpa')
```

```python
# Search for jurisprudence
dados_cjsg = tjpa.cjsg('dano moral', paginas=range(1,3))

print(dados_cjsg.shape)
dados_cjsg.head(3)
```

## Available columns

```python
dados_cjsg.columns.tolist()
```

## Preview ementa

```python
print(dados_cjsg['ementa'].iloc[0][:300])
```

## Using filters

The TJPA scraper supports several filters: date ranges, relator, collegiate judging body, class, and subject.

```python
# Filter by judgment date range
dados_filtrados = tjpa.cjsg(
    'dano moral',
    data_julgamento_inicio='2024-01-01',
    data_julgamento_fim='2024-06-30',
    paginas=1,
)

print(dados_filtrados.shape)
dados_filtrados.head(3)
```

## Download and parse separately

You can also download the raw responses and parse them in a second step.

```python
# Download raw JSON responses
brutos = tjpa.cjsg_download('dano moral', paginas=1)

# Inspect the raw response structure
print(f"Number of pages downloaded: {len(brutos)}")
print(f"Total results: {brutos[0]['data']['totalElements']}")
print(f"Results in first page: {len(brutos[0]['data']['content'])}")
```

```python
# Parse raw responses into a DataFrame
df_parsed = tjpa.cjsg_parse(brutos)

print(df_parsed.shape)
df_parsed.head(3)
```

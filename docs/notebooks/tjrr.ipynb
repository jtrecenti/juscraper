# TJRR - Tribunal de Justica de Roraima

Jurisprudence scraper for TJRR, via JSF/PrimeFaces with ViewState management.

| Feature | Available |
|---------|----------|
| cjsg    | Yes      |

```python
import juscraper as jus

tjrr = jus.scraper('tjrr')
```

## Basic search

```python
df = tjrr.cjsg('dano moral', paginas=range(1,3))

print(df.shape)
df.head(3)
```

## Available columns

```python
brutos = tjrr.cjsg_download('dano moral', paginas=1)
df_parsed = tjrr.cjsg_parse(brutos)

print(df_parsed.shape)
df_parsed.head(3)
```

## Preview ementa

## Using filters

Available filters: `relator`, `orgao_julgador` (list of codes), `especie` (list of resource types).

```python
df_filtered = tjrr.cjsg(
    'direito',
    data_inicio='2024-01-01',
    data_fim='2024-06-30',
    paginas=1,
)

print(df_filtered.shape)
df_filtered.head(3)
```

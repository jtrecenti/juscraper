# TJRO - Tribunal de Justica de Rondonia

Jurisprudence scraper for TJRO, via Elasticsearch API (JURIS).

| Feature | Available |
|---------|----------|
| cjsg    | Yes      |

```python
import juscraper as jus

tjro = jus.scraper('tjro')
```

## Basic search

```python
df = tjro.cjsg('dano moral', paginas=1)

print(df.shape)
df.head(3)
```

## Available columns

```python
brutos = tjro.cjsg_download('dano moral', paginas=1)
df_parsed = tjro.cjsg_parse(brutos)

print(df_parsed.shape)
df_parsed.head(3)
```

## Preview ementa

```python
print(df['ementa'].iloc[0][:300])
```

## Using filters

Available filters: `tipo` (list), `nr_processo`, `magistrado`, `orgao_julgador`, `orgao_julgador_colegiado`, `classe_judicial`, `instancia`, `termo_exato` (bool).

```python
df_filtered = tjro.cjsg(
    'direito',
    data_inicio='2024-01-01',
    data_fim='2024-06-30',
    paginas=1,
)

print(df_filtered.shape)
df_filtered.head(3)
```

# TJRN - Tribunal de Justica do Rio Grande do Norte

Jurisprudence scraper for TJRN, via Elasticsearch API.

| Feature | Available |
|---------|----------|
| cjsg    | Yes      |

```python
import juscraper as jus

tjrn = jus.scraper('tjrn')
```

## Basic search

```python
df = tjrn.cjsg('dano moral', paginas=range(1, 3))

print(df.shape)
df.head(3)
```

## Available columns

```python
brutos = tjrn.cjsg_download('dano moral', paginas=1)
df_parsed = tjrn.cjsg_parse(brutos)

print(df_parsed.shape)
df_parsed.head(3)
```

## Preview ementa

## Using filters

Available filters: `nr_processo`, `id_classe_judicial`, `id_orgao_julgador`, `id_relator`, `id_colegiado`, `sistema` (PJE/SAJ), `decisoes`, `jurisdicoes`, `grau`.

```python
df_filtered = tjrn.cjsg(
    'direito',
    data_inicio='2024-01-01',
    data_fim='2024-06-30',
    paginas=1,
)

print(df_filtered.shape)
df_filtered.head(3)
```

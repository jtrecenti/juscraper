# TJBA - Tribunal de Justica do Estado da Bahia

Scraper for the jurisprudence search of TJBA.

| Feature | Available |
|---------|----------|
| cjsg    | Yes      |
| cpopg   | No       |
| cposg   | No       |

## Basic search

```python
import juscraper as jus

tjba = jus.scraper('tjba')

df = tjba.cjsg('dano moral', paginas=1)
print(df.shape)
df.head(3)
```

## Using filters

TJBA supports filtering by publication date range.

```python
df_filtrado = tjba.cjsg(
    'direito',
    data_publicacao_inicio='2024-01-01',
    data_publicacao_fim='2024-06-30',
    paginas=1,
)
print(df_filtrado.shape)
df_filtrado.head(3)
```

## Filtering by decision type

You can filter by instance (2nd degree, turmas recursais) and decision type (acordaos, monocratic decisions).

```python
# Only acordaos from 2nd degree
df_acordaos = tjba.cjsg(
    'direito',
    segundo_grau=True,
    turmas_recursais=False,
    tipo_acordaos=True,
    tipo_decisoes_monocraticas=False,
    paginas=1,
)
print(df_acordaos.shape)
df_acordaos['tipo_decisao'].value_counts()
```

## Download and parse separately

```python
# Download raw results
brutos = tjba.cjsg_download('direito', paginas=1)

# Inspect raw response structure
print(type(brutos), len(brutos))
print(brutos[0]['data']['filter'].keys())

# Parse into DataFrame
df_parsed = tjba.cjsg_parse(brutos)
df_parsed.head(3)
```

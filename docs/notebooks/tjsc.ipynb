# TJSC - Tribunal de Justica de Santa Catarina

Jurisprudence scraper for TJSC, via the eproc PHP system.

| Feature | Available |
|---------|----------|
| cjsg    | Yes      |

```python
import juscraper as jus

tjsc = jus.scraper('tjsc')
```

## Basic search

```python
df = tjsc.cjsg('dano moral', paginas=range(1, 3))

print(df.shape)
df.head(3)
```

## Available columns

```python
brutos = tjsc.cjsg_download('dano moral', paginas=1)
df_parsed = tjsc.cjsg_parse(brutos)

print(df_parsed.shape)
df_parsed.head(3)
```

## Preview ementa

```python
if 'ementa' in df.columns and len(df) > 0:
    print(df['ementa'].iloc[0][:300])
```

## Using filters

Available filters: `campo` (E=ementa, I=inteiro teor), `processo`, `data_decisao_inicio/fim`, `data_publicacao_inicio/fim`.

```python
df_filtered = tjsc.cjsg(
    'direito',
    campo='E',
    paginas=1,
)

print(df_filtered.shape)
df_filtered.head(3)
```

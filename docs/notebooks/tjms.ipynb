# TJMS - Tribunal de Justica de Mato Grosso do Sul

Scraper for the jurisprudence search (CJSG) of the Tribunal de Justica de Mato Grosso do Sul.

| Feature | Available |
|---------|----------|
| cjsg    | Yes      |
| cpopg   | No       |
| cposg   | No       |

## Basic search

```python
import juscraper as jus

tjms = jus.scraper('tjms')
```

```python
df = tjms.cjsg('dano moral', paginas=range(1, 3))
print(df.shape)
df.head(3)
```

## Using filters

You can filter by judgment date, publication date, decision type, and other fields.

```python
df_filtered = tjms.cjsg(
    'direito',
    data_julgamento_inicio='01/01/2024',
    data_julgamento_fim='30/06/2024',
    paginas=1,
)
print(df_filtered.shape)
df_filtered.head(3)
```

## Download and parse separately

You can also download the raw HTML files and parse them in a separate step.

```python
import tempfile, os

tmp = tempfile.mkdtemp()
pasta = tjms.cjsg_download('direito', paginas=1, diretorio=tmp)
print('Downloaded to:', pasta)
print('Files:', os.listdir(pasta))
```

```python
df_parsed = tjms.cjsg_parse(pasta)
print(df_parsed.shape)
df_parsed.head(3)
```

# TJPI - Tribunal de Justica do Piaui

Jurisprudence scraper for TJPI, via HTML parsing from the JusPI platform.

| Feature | Available |
|---------|----------|
| cjsg    | Yes      |

```python
import juscraper as jus

tjpi = jus.scraper('tjpi')
```

## Basic search

```python
df = tjpi.cjsg('dano moral', paginas=range(1, 3))

print(df.shape)
df.head(3)
```

## Available columns

```python
brutos = tjpi.cjsg_download('dano moral', paginas=1)
df_parsed = tjpi.cjsg_parse(brutos)

print(df_parsed.shape)
df_parsed.head(3)
```

## Preview ementa

```python
print(df['ementa'].iloc[0][:300])
```

## Using filters

Available filters: `tipo` (acordao, decisao_terminativa, sumula), `relator`, `classe`, `orgao`.

```python
df_filtered = tjpi.cjsg('direito', tipo='acordao', paginas=1)

print(df_filtered.shape)
df_filtered.head(3)
```

# TJPE - Tribunal de Justica de Pernambuco

Scraper for jurisprudence searches on the TJPE website.

| Feature | Available |
|---------|-----------|
| `cjsg`  | Yes       |
| `cpopg` | No        |
| `cposg` | No        |

## Basic search

```python
import juscraper as jus

tjpe = jus.scraper('tjpe')

df = tjpe.cjsg('dano moral', paginas=1)
print(df.shape)
df.head(3)
```

## Available columns

```python
df.columns.tolist()
```

## Using date filters

```python
# Date filters use DD/MM/YYYY format
df_filtered = tjpe.cjsg(
    'direito',
    data_julgamento_inicio='01/01/2024',
    data_julgamento_fim='30/06/2024',
    paginas=1,
)
print(df_filtered.shape)
df_filtered[['processo', 'relator', 'data_julgamento']].head(3)
```

## Download and parse separately

```python
# Download raw HTML pages
raw = tjpe.cjsg_download('direito', paginas=1)
print(f"Downloaded {len(raw)} page(s)")
print(f"Raw HTML size: {len(raw[0])} bytes")

# Parse into DataFrame
df_parsed = tjpe.cjsg_parse(raw)
print(f"Parsed {len(df_parsed)} results")
df_parsed.head(3)
```

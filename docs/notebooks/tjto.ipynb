# TJTO - Tribunal de Justica do Estado do Tocantins

Scraper for the jurisprudence search of TJTO.

| Feature | Status |
|---------|--------|
| cjsg (Acordaos - 2nd instance) | Supported |
| cjsg (Decisoes Monocraticas - 2nd instance) | Supported |
| cjsg (Sentencas - 2nd instance) | Supported |
| cjpg (Acordaos - 1st instance) | Supported |
| cjpg (Decisoes Monocraticas - 1st instance) | Supported |
| cjpg (Sentencas - 1st instance) | Supported |
| cpopg | Not implemented |
| cposg | Not implemented |

## Basic search (2nd instance)

```python
import juscraper as jus

tjto = jus.scraper('tjto')

dados_cjsg = tjto.cjsg('dano moral', paginas=range(1, 3))
print(dados_cjsg.shape)
dados_cjsg.head(3)
```

## Available columns

```python
dados_cjsg.columns.tolist()
```

## Preview result fields

```python
row = dados_cjsg.iloc[0]
print(f"Processo: {row['processo']}")
print(f"Classe: {row['classe']}")
print(f"Relator: {row['relator']}")
print(f"Competencia: {row['competencia']}")
print(f"Data Julgamento: {row['data_julgamento']}")
```

## Using filters

You can filter by judgment date range and sort order.

```python
dados_filtrados = tjto.cjsg(
    'direito',
    paginas=range(1, 3),
    data_julgamento_inicio='01/01/2025',
    data_julgamento_fim='30/06/2025',
)
print(dados_filtrados.shape)
dados_filtrados.head(3)
```

## Querying different document types

Besides `acordaos` (default), you can also query `decisoes` (monocratic decisions) and `sentencas` (sentences).

```python
dados_decisoes = tjto.cjsg('direito', paginas=range(1, 3), tipo_documento='decisoes')
print(dados_decisoes.shape)
dados_decisoes.head(3)
```

## Download and parse separately

You can also download the raw HTML first and parse it later.

```python
brutos = tjto.cjsg_download('direito', paginas=range(1, 3))
print(f"Pages downloaded: {len(brutos)}")
print(f"Raw HTML size (page 1): {len(brutos[0])} chars")

dados = tjto.cjsg_parse(brutos)
print(f"\nParsed shape: {dados.shape}")
dados.head(3)
```

## First instance search (cjpg)

Use `cjpg()` to search first-instance jurisprudence. It accepts the same parameters as `cjsg()`.

```python
dados_cjpg = tjto.cjpg('dano moral', paginas=range(1, 3))
print(dados_cjpg.shape)
dados_cjpg.head(3)
```

### First instance with filters

```python
dados_cjpg_sentencas = tjto.cjpg(
    'direito',
    paginas=range(1, 3),
    tipo_documento='sentencas',
)
print(dados_cjpg_sentencas.shape)
dados_cjpg_sentencas.head(3)
```
